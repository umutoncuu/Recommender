```python
import pandas as pd
import numpy as np

path = "C:\\Users\\hp\\Desktop\\DataCamp Data\\movies.csv"

data = pd.read_csv(path)
data.head(5)
```

```python
movie_genres = ['Action','Adventure','Animation','Children','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','IMAX','Romance','Sci-Fi','Thriller','War','Western']
```

```python
for i in movie_genres:
    data["{}".format(i)] = np.zeros(len(data["genres"]), dtype=np.int8)
```

```python
for i in movie_genres:
    for x in range(len(data["genres"])):
        if i in (data["genres"][x].split("|")):
            data["{}".format(i)][x] = 1
        else:
            data["{}".format(i)][x] = 0
```

```python
data
```

```python
data = data.drop(["movieId", "genres"], axis=1)
```

```python
data.index = data["title"]
```

```python
data = data.drop("title", axis=1)
```

```python
data
```

```python
from scipy.spatial.distance import pdist, squareform
jaccard_distances = pdist(data.values, metric="jaccard")
square_jaccard = squareform(jaccard_distances)
square_jaccard
```

```python
jaccard_similarity = 1 - square_jaccard
jaccard_similarity
```

```python
item_item_matrix = pd.DataFrame(jaccard_similarity,
                          index=data.index,
                          columns=data.index)
```
